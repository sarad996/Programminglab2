### **1. Distribuzioni di probabilità**
Genera 1000 valori casuali da una distribuzione gamma con parametro di forma pari a 1.
Suggerimento: il parametro di forma viene passato come primo argomento quando si "congela" la distribuzione.

Traccia l’istogramma del campione e sovrapponi la PDF della distribuzione.

Stima il parametro di forma dal campione usando il metodo fit.

Extra:
Le distribuzioni hanno molti metodi utili. Esplorali usando il completamento automatico con il tasto TAB.

Traccia la funzione di distribuzione cumulativa (CDF).

Calcola la varianza.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp

# generazione di 1.000 valori casuali da una gamma di forma 1
gamma_dist = sp.stats.gamma(a=1)
campioni = gamma_dist.rvs(size=1000)

# plot dell'istogramma con la pdf della distribuzione
plt.figure(figsize=(10,6))

plt.hist(campioni, bins=50, density=True, label='Istogramma del campione della gamma', color='blue', alpha=0.6)
x = np.linspace(0,np.max(campioni), 500)
plt.plot(x, gamma_dist.pdf(x), label='PDF', color='green')

plt.title('Distribuzione Gamma (forma=1)')
plt.xlabel('Valori del campione')
plt.ylabel('Densità')
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()

# stima dei parametri della distribuzione Gamma del campione
a, loc, scale = sp.stats.gamma.fit(campioni)

print('parametri stimati:')
print(f'- a = {a}\n- loc = {loc}\n- scale = {scale}')

# distribuzione cumulativa CDF
gamma_cdf = gamma_dist.cdf(x)

plt.figure(figsize=(10,6))

plt.plot(x, gamma_cdf, label='CDF', color='blue')

plt.title('Funzione di ripartizione CDF')
plt.xlabel('Valori x')
plt.tight_layout()
plt.legend()
plt.grid(True)
plt.show()

# calcolo della varianza campionaria
var = np.var(campioni, ddof=1)
print(f'Varianza campionaria: {var}')
```

### **2. Fitta i dati**
Prova a fittare i dati sottostante con le migliori curve, calcola il MAE e l'RMSE

```python
import numpy as np
import matplotlib.pyplot as plt

temp_max = np.array([17, 19, 21, 28, 33, 38, 37, 37, 31, 23, 19, 18])
temp_min = np.array([-62, -59, -56, -46, -32, -18, -9, -13, -25, -46, -52, -58])

months = np.arange(12)

plt.plot(months, temp_max, "ro")
plt.plot(months, temp_min, "bo")
plt.xlabel("Month")
plt.ylabel("Min and max temperature")
plt.show()
```

```python
# fit polinomiale di grado 3
coef_max = np.polyfit(months, temp_max, 3)
coef_min = np.polyfit(months, temp_min, 3)

# predizioni
poly_max = np.poly1d(coef_max)
poly_min = np.poly1d(coef_min)
pred_max = poly_max(months)
pred_min = poly_min(months)

# calcolo MAE e RMSE
mae_max = np.mean(np.abs(temp_max-pred_max))
rmse_max = np.sqrt(np.mean((temp_max-pred_max)**2))
mae_min = np.mean(np.abs(temp_min - pred_min))
rmse_min = np.sqrt(np.mean((temp_min - pred_min)**2))

# plot
plt.figure(figsize=(10,6))

plt.plot(months, temp_max, "ro", alpha=0.5, label='Temp max osservate')
plt.plot(months, temp_min, "bo", alpha=0.5, label='Temp min osservate')

plt.plot(months, pred_max, color='r', linestyle='--', label='Fit max')
plt.plot(months, pred_min, color='b', linestyle='--', label='Fit min')

plt.title('Fitting polinomiale di grado 3')
plt.xlabel('Months')
plt.ylabel('Min/max temperature')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Stampa dei risultati
print("Temperature massime:")
print(f"-  MAE = {mae_max}")
print(f"-  RMSE = {rmse_max}")
print("Temperature minime:")
print(f"-  MAE = {mae_min}")
print(f"-  RMSE = {rmse_min}")

# fit polinomiale di grado 5
coef_max = np.polyfit(months, temp_max, 5)
coef_min = np.polyfit(months, temp_min, 5)

# predizioni
poly_max = np.poly1d(coef_max)
poly_min = np.poly1d(coef_min)
pred_max = poly_max(months)
pred_min = poly_min(months)

# calcolo MAE e RMSE
mae_max = np.mean(np.abs(temp_max-pred_max))
rmse_max = np.sqrt(np.mean((temp_max-pred_max)**2))
mae_min = np.mean(np.abs(temp_min - pred_min))
rmse_min = np.sqrt(np.mean((temp_min - pred_min)**2))

# plot
plt.figure(figsize=(10,6))

plt.plot(months, temp_max, "ro", alpha=0.5, label='Temp max osservate')
plt.plot(months, temp_min, "bo", alpha=0.5, label='Temp min osservate')

plt.plot(months, pred_max, color='r', linestyle='--', label='Fit max')
plt.plot(months, pred_min, color='b', linestyle='--', label='Fit min')

plt.title('Fitting polinomiale di grado 5')
plt.xlabel('Months')
plt.ylabel('Min/max temperature')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Stampa dei risultati
print("Temperature massime:")
print(f"-  MAE = {mae_max}")
print(f"-  RMSE = {rmse_max}")
print("Temperature minime:")
print(f"-  MAE = {mae_min}")
print(f"-  RMSE = {rmse_min}")
```
